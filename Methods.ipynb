```python
from typing import Callable
import math

func = lambda x : math.cos(x) - math.sin(x)

def Bisection(a: float, b: float, func: Callable[[float], float], iterations: int = 1, tolerance: float = 10**-6) -> None:
    a_i = a
    b_i = b

    for _ in range(1, iterations):
        x_i = (a_i + b_i) / 2

        if func(x_i) == 0 or ((b_i - a_i) / 2) < tolerance: 
            break

        if func(a_i)*func(x_i) < 0: b_i = x_i

        if func(x_i)*func(b_i) < 0: a_i = x_i
    
    print((a_i + b_i) / 2)

Bisection(0, 1, func, 20)
```

```python
x_guesses = [0.9, -0.7, 1.2, 2, 1.3, -1]
answers = [5/2, 3/2, 1, 1, 3/2, 5/2]
matrix = [[3, -1, 0, 0, 0, 1/2],
          [-1, 3, -1, 0, 1/2, 0],
          [0, -1, 3, -1, 0, 0],
          [0, 0, -1, 3, -1, 0],
          [0, 1/2, 0, -1, 3, -1],
          [1/2, 0, 0, 0, -1, 3]]

def Jacobi(x_guesses: list, answers: list, matrix: list[list], iterations: int = 1) -> None:
    N = len(answers)
    x_est = x_guesses

    for _ in range(1, iterations):
        x_new = [0] * N
        
        for eq_ind in range(N):
            res = answers[eq_ind]

            for e_ind in range(N):
                if e_ind != eq_ind:
                    res -= matrix[eq_ind][e_ind]*x_est[e_ind]

            x_new[eq_ind] = res / matrix[eq_ind][eq_ind]
        x_est = x_new

        print(x_est)

Jacobi(x_guesses, answers, matrix, 50)
```

```python
import numpy as np

def LeastSquares(answers: list[int], estimateTo: int, order: int = 1):
    l = len(answers)
    matrixSTR = ''
    for index in range(l):
        matrixSTR += '1'
        for o in range(1, order+1):
            matrixSTR += f' {index**o}'
        matrixSTR += '; '

    matrixSTR = matrixSTR.strip().rstrip(';')

    A = np.matrix(matrixSTR)
    AT = A.transpose()
    ATA = AT @ A
    print(ATA)
    
    b = np.array(answers)
    ATb = AT @ b
    print(ATb)

    coefficients = np.linalg.solve(ATA, ATb.transpose())

    print(coefficients)

    sum = 0
    for ord in range(order+1):
        sum += coefficients[ord]*estimateTo**ord

    print(sum)
    
LeastSquares([67052,68008,69809,72024,73400,72063,74669,74487,74065,76777], 26, 3)
```
